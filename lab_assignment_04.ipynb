(3) Using factorplot to show the relationship between numerical and categorical value. Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:

```python
# Question 2.1)
import pandas as pd
url = 'https://raw.githubusercontent.com/suthapalliuday/INFO5502-Summer2022/main/datasets/lab-assignment-04-student-performance.csv'
# reading raw data from git
df = pd.read_csv(url)
for col in df.columns:
    print("Column -",col)
    print("Number of rows -", df[col].shape[0])
    print("NUmber of unique values in this column -", df[col].value_counts().shape[0])
    print("Frequency of each element:")
    print(df[col].value_counts())
    print()
    print()
```

```python
# Question 2.2)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
ax=sns.countplot(x='sh',data=df)
plt.rcParams['figure.figsize']=[10,10]
plt.xlabel("Study Hours")
plt.ylabel("No. of students")
plt.show()
```

```python
# Question 2.3)
sns.catplot(x = "sh", y = "esp", data = df,kind="strip")
plt.xlabel("Study Hours")
plt.ylabel("End Semester Percentage")
plt.show()
```

```python
# Question 2.4)
for x in df.columns:
    df[x] = df[x].astype('category')
    cat_columns = df.select_dtypes(['category']).columns
    df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
corr=df.corr()

sns.set(rc={'figure.figsize':(30,15)})
sns.heatmap(corr,annot=True)
```
